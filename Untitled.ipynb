```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Data for plotting
mass = np.array([100.46, 150.32, 200.15, 249.39, 299.22])  # Mass in grams
period_squared = np.array([0.67, 0.83, 1.04, 1.28, 1.49])  # Period squared in s^2
rmsd_period_squared = np.array([0.03, 0.02, 0.03, 0.07, 0.03])  # RMSD for period squared

# Define a linear function for fitting
def linear_func(x, a, b):
    return a * x + b

# Perform curve fitting to find the best-fit line and uncertainties in the slope
popt, pcov = curve_fit(linear_func, mass, period_squared, sigma=rmsd_period_squared, absolute_sigma=True)
slope, intercept = popt
slope_uncertainty = np.sqrt(pcov[0, 0])

# Generate data for the line of best fit
mass_fit = np.linspace(min(mass), max(mass), 100)
period_squared_fit = linear_func(mass_fit, *popt)

# Plotting
plt.figure(figsize=(8, 6))
plt.errorbar(mass, period_squared, yerr=rmsd_period_squared, fmt='o', label='Data points', capsize=5)
plt.plot(mass_fit, period_squared_fit, 'r-', label=f'Best fit line: y = {slope:.3f}x + {intercept:.3f}')
plt.xlabel('Total Load Mass (g)')
plt.ylabel('Period Squared (s^2)')
plt.title('Period Squared vs Total Load Mass')
plt.legend()
plt.grid(True)
plt.show()
```
